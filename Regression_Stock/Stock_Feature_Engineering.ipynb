# Part 2 Feature Engineering

- Price Features - Feature Engineering

```python
# Daily return
df['daily_return'] = (df['close'] - df['open']) / df['open'] 
df.head(5)
```

```python
# High - Low Spread
df['high_low_spread'] = (df['high'] - df['low']) / df['open']
df.head(5)
```

```python
# Open-Close Spread
df['open_close_spread'] = (df['close'] - df['open']) / df['open']
df.head(5)
```

```python
# Rolling mean for N days
N = 5 # or 10, 20, 30
df[f'rolling_mean_{N}'] = df['close'].rolling(N).mean() 
df.head(5)
```

```python
# Rolling Standard Deviation (N Days)
df[f'rolling_standard_deviation_{N}'] = df['close'].rolling(N).std()
df.head(5)
```

```python
# Cumulative return
df['cumulative_return'] = df['close'] / df['close'].iloc[0] - 1
df.head(5)
```

```python
# Log return
df['log_return'] = np.log(df['close'] / df['close'].shift(1))
df.head(5)
```

```python
# Rolling Min and Rolling Max
df[f'rolling_max_{N}'] = df['close'].rolling(N).max()
df[f'rolling_min_{N}'] = df['close'].rolling(N).min()
df.head(5)
```

```python
# Price momentum
df[f'momentum_{N}'] = df['close'] - df['close'].shift(N)
df.head(5)
```

```python
# Z- Score
df[f'z_score_{N}'] = (df['close'] - df[f'rolling_mean_{N}']) / df[f'rolling_standard_deviation_{N}']
df.head(5)
```

```python
# LAG Feature
M = 5
for lag in range(1, M + 1):
    df[f'close_lag_{lag}'] = df['close'].shift(lag)
df.head(5)
```

```python
# Calculate the Bolliger Bands Width
rolling_mean = df['close'].rolling(N).mean()
rolling_std = df['close'].rolling(N).std()
df['bollinger_upper'] = rolling_mean + 2 * rolling_std
df['bollinger_lower'] = rolling_mean - 2 * rolling_std
df['bollinger_width'] = (df['bollinger_upper'] - df['bollinger_lower']) / rolling_mean
df.head(5)
```

## Volume Based Feature Engineering

```python
# Volume Spike (Compare it to 5day rolling average)
df['volume_spike'] = df['volume'] / df['volume'].rolling(5).mean()
df.head(5)
```

```python
# Create also volume spike alert if the volume is 2 times more than the 5-day average 
df['volume_spike_alert'] = df['volume_spike'] > 2.0
df.head(2)
```

# Momentum Indicators

```python
# Rate Of Change
# the presence of NaN in your Rate of Change (ROC) is expected for the first N rows (where N is the lookback period you used, such as 5 days), 
N = 5 # days
df['ROC'] = df.groupby('ticker')['close'].transform(
    lambda x: (x - x.shift(N) / x.shift(N) )
)
df.head(100)
```

```python
# Relative Strength Index (RSI), default lookback period 14
def compute_rsi_index(series, period = 14):
    delta = series.diff() # Calculate the difference between consecutive prices
    gain = np.where(delta > 0, delta, 0) # Separate gains (positive changes); set other values to 0
    loss = np.where(delta < 0, -delta, 0) # Separate losses (negative changes); invert and set other values to 0
    average_gain = pd.Series(gain).rolling(window = period).mean() # Compute the rolling average gain over the given period
    average_loss = pd.Series(loss).rolling(window = period).mean() # Compute the rolling average loss over the given period
    rs = average_gain / average_loss # Calculate the Relative Strength (RS) = average gain / average loss
    rsi = 100 - (100 / (1 - rs)) # Compute the RSI using the RS value
    return pd.Series(rsi) # Return the RSI as a Pandas Series

df['RSI_14'] = df.groupby('ticker')['close'].transform(
    lambda x: compute_rsi_index(x, period = 14)
)

print(df[df['RSI_14'].notna()])
```

- MACD and Signal Line

```python
df['EMA_12'] = df.groupby('ticker')['close'].transform(
    lambda x: x.ewm(span=12, adjust=False).mean()
)
```

```python
df['EMA_26'] = df.groupby('ticker')['close'].transform(
    lambda x: x.ewm(span=26, adjust=False).mean()
)
```

```python
df['MACD'] = df['EMA_12'] - df['EMA_26']
```

```python
df['MACD_signal'] = df.groupby('ticker')['MACD'].transform(
    lambda x: x.ewm(span=9, adjust=False).mean()
)
```

# Trend Indicators

- Create the Moving Averages
- SMA
- EMA 
- Window = For how many days

```python
window = 5
df['SMA'] = df['close'].rolling(window = window).mean()
df['EMA'] = df['close'].ewm(span = window, adjust= False).mean()
df.head(5) 
```

- Binary Indicators
- Price_Above_SMA
- Price_Abone_EMA

```python
df['Price_Above_SMA'] = (df['close'] > df['SMA']).astype(int)
df['Price_Above_EMA'] = (df['close'] > df['EMA']).astype(int)
df.head(5)
```

- Slope of the moving averages

```python
df['SMA_Slope'] = df['SMA'].diff()
df['EMA_Slope'] = df['EMA'].diff()
df.head(5)
```

- Trend Strength Categories

```python
# Trend Strength Categories
def categorize_trend(slope):
    if slope > 0.05:
        return "Strong Uptrend"
    elif slope > 0:
        return "Weak Uptrend"
    elif slope < -0.05:
        return "Strong Downtrend"
    elif slope < 0:
        return "Weak Downtrend"
    else:
        return "Flat"
```

```python
# Implement the Trend Strength Categories
df["Trend_Strength"] = df["SMA_Slope"].apply(categorize_trend)
df.head(5)
```

- Calculate also the MACD_Hist

```python
df["MACD_Hist"] = df["MACD"] - df["MACD_signal"]
df.head(5)
```

# Create the Buy/Sell Signals

- RSI Strategy: Buy when RSI < 30, Sell when RSI > 70

```python
df['RSI_Signal'] = np.where( df["RSI_14"] < 30, "Buy",
                     np.where(df["RSI_14"] > 70, "Sell", "Hold"))
df.head(10)
```

- MACD Strategy: Buy when MACD crosses above signal line, Sell when it crosses below

```python
df['MACD_Signal_Flag'] = np.where(
    (df['MACD'] > df['MACD_signal']) & (df['MACD'].shift(1) <= df['MACD_signal'].shift(1)),
    "Buy",
    np.where(
        (df['MACD'] < df['MACD_signal']) & (df['MACD'].shift(1) >= df['MACD_signal'].shift(1)),
        "Sell",
        "Hold"
    )
)
df.head(5)
```

```python
# Doji pattern
df["Doji"] = np.where(abs(df["close"] - df["open"]) <= (df["high"] - df["low"]) * 0.1, 1, 0)
df.head(5)
```

```python
# Detect Hammer
body = abs(df["close"] - df["open"])
lower_shadow = df["open"].where(df["close"] > df["open"], df["close"]) - df["low"]
upper_shadow = df["high"] - df["close"].where(df["close"] > df["open"], df["open"])
df["Hammer"] = np.where(
    (body <= (df["high"] - df["low"]) * 0.3) &  # small body
    (lower_shadow >= 2 * body) &               # long lower wick
    (upper_shadow <= body),                    # short/absent upper wick
    1, 0
)
df.head(5)
```

```python
# Detect Bullish
df["Bullish_Engulfing"] = np.where(
    (df["close"] > df["open"]) &  # current candle is green
    (df["close"].shift(1) < df["open"].shift(1)) &  # previous candle was red
    (df["open"] < df["close"].shift(1)) &  # today's open < yesterday's close
    (df["close"] > df["open"].shift(1)),   # today's close > yesterday's open
    1, 0
)
```

```python
# Detect Morning Star (Simplified 3-bar pattern) ---
df["Morning_Star"] = np.where(
    (df["close"].shift(2) < df["open"].shift(2)) &  # 1st: red
    (abs(df["close"].shift(1) - df["open"].shift(1)) < (df["high"].shift(1) - df["low"].shift(1)) * 0.3) &  # 2nd: small body
    (df["close"] > df["open"]) & (df["close"] > df["close"].shift(2)),  # 3rd: green and closes above 1st
    1, 0
)
```

```python
# Detect Evening Star
df["Evening_Star"] = np.where(
    (df["close"].shift(2) > df["open"].shift(2)) &  # 1st: green
    (abs(df["close"].shift(1) - df["open"].shift(1)) < (df["high"].shift(1) - df["low"].shift(1)) * 0.3) &  # 2nd: small body
    (df["close"] < df["open"]) & (df["close"] < df["close"].shift(2)),  # 3rd: red and closes below 1st
    1, 0
)
```
